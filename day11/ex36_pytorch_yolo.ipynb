### 8. 동영상 플레이하면서 물체인식

```python
import cv2
import time
from ultralytics import YOLO

# 1. 모델 로드
model = YOLO('yolo11n.pt')

# 2. 웹캠 연결
cap = cv2.VideoCapture(0)

# 초기 FPS 계산용 시간 설정 (0으로 나누기 방지)
prev_time = time.time()

while True:
    # 프레임 읽기
    ret, frame = cap.read()
    if not ret: 
        break  # 더이상 영상(프레임)이 없으면 종료

    # 3. YOLO 영상 추론 (verbose=False로 터미널을 깔끔하게 유지)
    results = model(frame, conf=0.5, verbose=False)
    pred_frame = results[0].plot()
    
    # 4. FPS 계산 (시간 차이가 0보다 클 때만 계산하여 안전성 확보)
    curr_time = time.time()
    time_diff = curr_time - prev_time
    fps = 1 / time_diff if time_diff > 0 else 0
    prev_time = curr_time

    # 5. [추가] FPS를 결과 화면(pred_frame) 좌측 상단에 초록색 글씨로 출력
    cv2.putText(
        pred_frame, 
        f"FPS: {int(fps)}", 
        (20, 50), 
        cv2.FONT_HERSHEY_SIMPLEX, 
        1, 
        (0, 255, 0), 
        2
    )
    
    # 6. [수정] 원본 frame 대신 YOLO 상자와 FPS가 포함된 pred_frame을 화면에 출력
    cv2.imshow('YOLO Player', pred_frame)

    # 'q' 키를 누르면 루프 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 7. 리소스 안전하게 해제
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)  # Jupyter 환경에서 창이 정상적으로 닫히도록 돕습니다.
```

### 9. 침입탐지

```python
import cv2
import time
from ultralytics import YOLO
import winsound

model = YOLO('yolo11n.pt')
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret: 
        break 

    # YOLO 영상 추론 (verbose=False를 넣으면 터미널이 깔끔해집니다)
    results = model(frame, conf=0.5, verbose=False)
    detected = False

    # 사람 인식 확인 루프
    for box in results[0].boxes:
        cls_id = int(box.cls[0])
        # [수정] 대시(-) 기호를 등호(=)로 변경하여 변수를 정상 정의합니다.
        label = model.names[cls_id]

        if label == 'person':
            detected = True
            break # 사람이 한 명이라도 발견되면 루프를 즉시 탈출하여 연산을 아낍니다.
    
    pred_frame = results[0].plot()

    if detected:
        cv2.putText(
            pred_frame,
            'Person Detected', # [수정] Persin -> Person 오타 수정
            (20, 50),
            cv2.FONT_HERSHEY_COMPLEX,
            1,
            (0, 0, 255),
            2
        )
        # 경고음 재생
        winsound.Beep(1000, 100) 

    cv2.imshow('Detection Alarm', pred_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 1. 필요한 라이브러리 가져오기
from ultralytics import YOLO
from huggingface_hub import hf_hub_download

# 2. Hugging Face에서 화재 감지 가중치 파일(.pt) 직접 다운로드하기
# 모델 페이지에 등록된 실제 파일명인 'firedetect-11s.pt'를 명시하여 가져옵니다.
model_path = hf_hub_download(repo_id="leeyunjai/yolo11-firedetect", filename="firedetect-11s.pt")
print(f"✅ 모델 다운로드 완료! 저장 위치: {model_path}")

# 3. 다운로드된 로컬 경로 파일로 YOLO 모델 로드
model = YOLO(model_path)

# 4. [수정] 올바른 이미지 주소 설정 (단순 웹사이트 메인이 아닌 이미지 파일 경로)
source = 'http://images.cocodataset.org/val2017/000000039769.jpg'

# 5. 이미지 추론 후 결과 저장
model.predict(source=source, save=True)
```

```python
# 5. [수정] 결과를 변수(results)에 저장하고 추론 수행
results = model.predict(source=source, save=True)

# 6. 결과 데이터 불러와서 출력하기
for result in results:
    boxes = result.boxes  # 검출된 상자(Bounding Box) 정보
    
    for box in boxes:
        # 클래스 번호 및 이름 불러오기
        cls_id = int(box.cls[0])
        label = model.names[cls_id]
        
        # 신뢰도 점수 (확률) 불러오기
        confidence = float(box.conf[0])
        
        # 사각형 좌표 불러오기 (좌상단 x, y, 우하단 x, y)
        xyxy = box.xyxy[0].tolist()
        
        print(f"🎯 검출된 물체: {label} (확률: {confidence:.2f})")
        print(f"📍 위치 좌표: {xyxy}")
```

```python
import cv2
import time
from ultralytics import YOLO
from huggingface_hub import hf_hub_download
import winsound

# 1. 화재 감지 전용 모델 로드
model_path = hf_hub_download(repo_id="leeyunjai/yolo11-firedetect", filename="firedetect-11s.pt")
model = YOLO(model_path)

# 2. 웹캠 켜기
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret: 
        break

    # 화재 감지 수행 (신뢰도 40% 이상만 검출)
    results = model(frame, conf=0.4, verbose=False)
    
    fire_detected = False
    
    # 검출된 물체들 확인
    for box in results[0].boxes:
        cls_id = int(box.cls[0])
        label = model.names[cls_id] # 커텀 모델의 라벨명 (fire 또는 smoke 등)
        
        # 모델이 정의한 화재 관련 키워드가 감지되었는지 확인
        if 'fire' in label.lower() or 'smoke' in label.lower():
            fire_detected = True
            break
            
    # 결과 상자 그리기
    pred_frame = results[0].plot()

    # 화재 발생 시 경고 시각화 및 비프음 발생
    if fire_detected:
        cv2.putText(
            pred_frame,
            '🔥 FIRE DETECTED 🔥',
            (20, 60),
            cv2.FONT_HERSHEY_COMPLEX,
            1.2,
            (0, 0, 255), # 빨간색 글씨
            3
        )
        winsound.Beep(2000, 200) # 높고 짧은 경고음

    # 화면 표시
    cv2.imshow('Fire Detection System', pred_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
