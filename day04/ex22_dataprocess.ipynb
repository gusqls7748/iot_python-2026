#### 결측치(Missing Value)
- 데이터 수집과정에서 누락되거나 측정되지 않은 빈 값
- 데이터분석에서는 걸측치를 처리하는 일이 가장 중요

````python
# 새 컬럼 만들기
df['수학_1'] = df['수학'] - 5
````

````python
#정렬
df.sort_value('수학_1')
````

````python
df.sort_values(by='수학_1', ascending=False)
````

````python
#걸측치 처리
df1 = df.dropna()
df1
````

````python
df2 = df.fillna(45)
df2
````

````python
#그룹화(GroupBy) - SQL이 Group By와 동일
df.groupby('수학').count()
````

````python
df[['국어', '영어', '수학']].fillna(45).groupby('수학').mean()
````

간단 예제

````python
data = {
    'name': ['유고', '요미', '철수', '예성'],
    'score': [80, 95, 60, 70]
}

df = pd.DataFrame(data)
df
````

````python
#CSV로 쓰게
df.to_csv('./scores.csv', encoding='utf-8')
````

````python
df['score'].mean()
````

````python
# 70점 이상만 추출
df['score'] >= 70
````

````python
df[df['score']>=70]
````

````python
# 합격여부
# 리스트 컴프리헨션으로 새 컬럼 생성
df['result'] = ['PASS' if s >= 70 else 'FAIL' for s in df['score']]
````

````python
df
````
