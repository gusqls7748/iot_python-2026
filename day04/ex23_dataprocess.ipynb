````python
# matplotlib 라이브러리 로드
import matplotlib.pyplot as plt
````

#### 기본 차트출력

````python
x = [1, 2, 3, 4, 5]    # x축 데이터
y = [10, 20, 15, 30, 25]  # y축 데이터

plt.plot(x, y)  # 라인차트
plt.show()      # 필요없는 차트설명 제외
````

#### 제목, 레이블 표시

- 기본적으로 한글 출력 안됨

````python
plt.plot(x, y)
plt.title('점수 추이')
plt.xlabel('시간')
plt.ylabel('Score')
plt.show()
````

- 한글 설정

````python
# 리소스인수, 폰트관리자, 리소스 포함
from matplotlib import rcParams, font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf'  # NanumGothic.ttf
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)  # 한글 설정
rcParams['axes.unicode_minus'] = False  # 마이너스 표시가 깨짐 방지
font
````

````python
plt.plot(x, y)
plt.title('점수 추이')      # 차트 제목
plt.xlabel('시간')          # x축 라벨
plt.ylabel('Score')         # y축 라벨
plt.show()
````

#### 포인트 모양 변경

````python
plt.plot(x, y, marker='d')  # 마커표시 *, ^, v, <, >, s, p, d
plt.title('점수 추이')      # 차트 제목
plt.xlabel('시간')          # x축 라벨
plt.ylabel('Score')         # y축 라벨
plt.show()
````

#### 라인, 색상, 포인트
- x, y 좌표 뒤에 문자열로 작성
    - 색상 :  b, g, r, c, m, y, k, w
    - 마커 : 위와 동일
    - 라인스타일 : -(line), --(dashed line), -.(dashed dot line), :(dotted line)

````python
plt.plot(x, y, 'm^--')      # 색상,마커, 라인을 조합
plt.title('점수 추이')      # 차트 제목
plt.xlabel('시간')          # x축 라벨
plt.ylabel('Score')         # y축 라벨
plt.show()
````

#### 스타일지정
- seaborn은 차트 생성전에 스타일 지정

````python
sns.set_style('darkgrid')   # 단순 스타일지정
sns.set_theme(font=font, rc={'axes.unicode_minus': False})  # 한글깨짐 방지
````

````python
plt.plot(x, y, 'm^--')      # 색상,마커, 라인을 조합
plt.title('점수 추이')      # 차트 제목
plt.xlabel('시간')          # x축 라벨
plt.ylabel('Score')         # y축 라벨
plt.show()
````

````python
# 막대 그래프
plt.bar(x, y)
plt.title('점수 추이')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
````

#### 차트종류
- plot() : 라인차트
- bar() : 막대그래프
- barh() : 가로막대그래프. x,y축 값을 변경
- hist() : 히스토그램. 분포표
- scatter() : 산점도. 데이터간의 관계 표현
- pie() : 파이차트. 비율표시

#### 차트 팁
- grid(True) : 맷플롯립만 사용시는 그리드 표시. Seaborn에서는 영향없음
- figure(figsize) : 그래프 크기 설정

````python
import numpy as np
````

````python
scores = list(np.random.randint(50, 100, 12))
list(scores)
````

````python
# 분포표
plt.figure(figsize=(10, 5))   # 차트 크기 설정 제일 먼저
plt.hist(scores, bins=5)
plt.title('학생점수 분포표')
plt.xlabel('점수대')
plt.ylabel('학생 수')
plt.show()
````

````python
# 막대 그래프
plt.figure(figsize=(10, 5))   # 차트 크기 설정
plt.bar(x, y)
plt.title('점수 추이')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
````

````python
# 산점도
study_time = [1,2,3,4,5,6,7,8]
score = [50, 55,65,72,78,81,93,98]

plt.figure(figsize=(10, 5))
plt.scatter(x=study_time, y=score, marker='^', color='g')
plt.title('학습시간 대비 점수 상관관계')
plt.xlabel('학습시간')
plt.ylabel('점수')
plt.show()
````

#### 3D 그래프

````python
# 주피터노트북 새로 생성한 뒤 첫번째 코드셸에 라이브러리 배치
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # 3D그래프용 클래스

import seaborn as sns
import numpy as np

# 리소스인수, 폰트관리자, 리소스 포함
from matplotlib import rcParams, font_manager, rc

font_path = 'C:/Windows/Fonts/NanumGothicCoding.ttf'  # NanumGothic.ttf
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)  # 한글 설정
rcParams['axes.unicode_minus'] = False  # 마이너스 표시가 깨짐 방지
````

````python
x = [1,2,3,4]
y = [10,20,15,30]
z = [5,10,8,12]

fig = plt.figure(figsize=(7, 7))   # 새로운 차트 선언해서 변수할당
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel('X축')
ax.set_ylabel('Y축')
ax.set_zlabel('Z축')

plt.title('세 축 데이터')
plt.show()
````

````python
# 3D Surface
x = np.linspace(-5, 5, 50)  # -5 ~ 5사이를 50개 간격으로 쪼갬
y = np.linspace(-5, 5, 50)  # -5 ~ 5사이를 50개 간격으로 쪼갬

x, y = np.meshgrid(x, y)
z = x**2 + y**2

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
# ax.plot_wireframe(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````
