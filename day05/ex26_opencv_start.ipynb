````python
# 이미지로드
img = cv2.imread('./cat.jpg')

cv2.imshow('origin', img) # origin 윈도우 제목으로 이미지 오픈
cv2.waitKey(0)        # 키보드 입력 대기, 무슨키든지 다 종료
cv2.destroyAllWindows() # 종료 후 열린 모든 윈도우창 종료
````

````python
# 동영상로드
cap = cv2.VideoCapture(filename='./beach.mp4')

while True: #무한루프
    ret, frame = cap.read() # FPS의 이미지 하나씩 읽어옴
    if not ret: break # 더이상 읽을 이미지 없으면 종료

    cv2.imshow('video', frame) # 이미지를 계속 변경
    if cv2.waitKey(1) == ord('q'): break    # 키보드 q입력하면 종료

cap.release()   # 메모리 해제
cv2.destroyAllWindows()
````

### 이미지 처리

````python
# 300x300 크기 빨강이미지 만들기
# 300(Width pixel), 300(height pixel), 3(channel: R G B)
img = np.zeros((300, 300, 3), dtype=np.uint8)
img[:] = (0, 0, 255) # OpenCV RGB를 거꾸로  사용 B, G, R

cv2.imwrite('./red_img.jpg', img)
````

````python
# 이미지 특정 위치를 변경
# imread 옵션 색상변경, 사이즈 변경
cat = cv2.imread(filename='./cat.jpg')

# 이미지 영역
# 고양이 이미지에서 특정 영역 지정
roi = cat[168:168 + 35, 318:318 + 192] # [행:행 + 높이, 열:열 + 넓이]
roi[:] = (0, 0, 0) # 검은색

cv2.imshow('cat parasite', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# imread 옵션
# flags
# IMREAD_COLOR, IMREAD_ANYCOLOR, IMREAD_COLOR_BGR = 색상 그대로
# IMREAD_COLOR_RGB =OpenCV 색상순서 거꾸로
# IMREAD_GRAYSCALE = 회색톤
# IMREAD_REDUCED_GRAYSCALE_2 = 회색톤 1/2 사이즈
# IMREAD_REDUCED_COLOR_2 = 컬러 1/2 사이즈
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_REDUCED_GRAYSCALE_2)

# 이미지 영역
# 고양이 이미지에서 특정 영역 지정
roi = cat[168:168 + 35, 318:318 + 192] # [행:행 + 높이, 열:열 + 넓이]
roi[:] = 0 # 0 # GRAYScale은 색이 하나 #(0,0,0) # 검은색

cv2.imshow('cat parasite', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 변경
# 로드한 이미지는 매트릭스 객체에 할당
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR)

# OpenCV 함수로 변경
gray = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY) # 컬라를 회색통으로 변경

cv2.imshow('origin', cat)
cv2.imshow('gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 채널 분리
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

(b, g, r) = cv2.split(cat)

# 채널 (B,G,R)를 단일 채널로 분리
# np.zeros() : 배열사이즈를 직접지정해서 0으로 배열초기화
# np.zeros_like(배열) : 지정된 배열사이즈를 0으로 초기화
blue = cv2.merge((b, np.zeros_like(b), np.zeros_like(b)))
green = cv2.merge((np.zeros_like(b), g, np.zeros_like(r)))
red = cv2.merge((np.zeros_like(b), np.zeros_like(r), r))


cv2.imshow('origin', cat)
cv2.imshow('blue', blue)
cv2.imshow('green',green)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 변경
# 로드한 이미지는 매트릭스 객체에 할당
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# OpenCV 함수로 변경
small = cv2.resize(cat, None, fx=0.5, fy=0.5) 

cv2.imshow('origin', cat)
cv2.imshow('1/2 size', small)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# OpenCV 함수로 변경
rot180 = cv2.rotate(cat, cv2.ROTATE_180)

cv2.imshow('origin', cat)
cv2.imshow('rot180', rot180)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#### OpenCV 이미지 처리 

- 주요기능 - 포토샵 등 이미지편집기 기능과 동일
    - 이미지 읽기, 저장, 출력
    - 이미지합성, 이미지 자르기
    - 색상변환
    - 크기변경
    - 회전, 이동
    - 이진화
    - 선명도 조절
    - 노이즈 제거
    - 대비/밝기 조절
    - 엣지 검출
    - 윤곽선 찾기
    - 퍼시펙티브 변환 - 원근감 이미지를 평탄화

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# 원그리기
# OpenCV 이미지 자체는 Mat(매트릭스)라서 행(height), 열(width) 순으로 처리
# 원 포인트는 (x,y)로 처리
# thickness=-1 내부 채우기
cv2.circle(cat, (416, 210), 100, (0, 255, 0), thickness=-1, lineType=cv2.LINE_AA)

cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# 텍스트
cv2.putText(cat, 'Kitty', (385, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 225), 5, cv2.LINE_AA)
cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ROI

- Region Of Interest : 관심영역
    - 2차원리스트(매트릭스, 행렬)
    - 영역추출법 - Mat[y시작점:y시작점+높이, x시작점:x시작점+넓이]

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# ROI선택
roi = cat[81:81+214, 295:295+240]

cv2.imshow('origin', cat)
cv2.imshow('roi', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat.jpg', flags=cv2.IMREAD_COLOR_BGR)

# 엣지(이미지 경계) 검출
# 그레이케일 변경
gray = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)

#canny 엣지
# threshold 임계값, 1은 약한 엣지, 2는 강한 엣지
canny = cv2.Canny(gray, threshold1=100, threshold2=200)

cv2.imshow('gray', gray)
cv2.imshow('canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
